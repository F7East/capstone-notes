# Adding Filters

```python
l_ha, l_hb, l_oiii = 6568.2, 4861, 5033.52
w_ha, w_hb, w_oiii = 93.88, 170, 59.36
```

```python
ha = [l_ha - w_ha/2, l_ha + w_ha/2]
hb = [l_hb - w_hb/2, l_hb + w_hb/2]
oiii = [l_oiii - w_oiii/2, l_oiii + w_oiii/2]
```

```python
plt.plot(x31, y31mjy, linewidth = 2)
plt.xlabel('Wavelength [$A$]')
plt.ylabel('Flux density [$mJy$]')
plt.axvline(x=ha[0], color='red', linestyle='--', linewidth = 1, label = 'ha')
plt.axvline(x=ha[1], color='red', linestyle='--', linewidth = 1)
plt.axvline(x=hb[0], color='green', linestyle='--', linewidth = 1, label = 'hb')
plt.axvline(x=hb[1], color='green', linestyle='--', linewidth = 1)
plt.axvline(x=oiii[0], color='orange', linestyle='--', linewidth = 1, label = 'oiii')
plt.axvline(x=oiii[1], color='orange', linestyle='--', linewidth = 1)
plt.legend()
plt.title('SP2031+410 standard star')
```
