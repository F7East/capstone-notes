```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
plt.show()
plt.style.use(['fivethirtyeight','seaborn-deep'])
```

```python
directory = '/Users/tengizibrayev/Desktop/capstone/line ratios'
```

```python
# Get a list of all files in the directory with a .dat extension
file_list = [f for f in os.listdir(directory) if f.endswith('.csv')]

dataframes = []
# Loop through the list of files
for csv_file in file_list:
    df = pd.read_csv(directory + '/' + csv_file)
    dataframes.append([df, csv_file])
```

```python
for i in dataframes[1:]:
#     try:
    df = i[0]
    # Create a histogram of the values in col2 that are floats
    df[df.columns[1]] = pd.to_numeric(df[df.columns[1]], errors='coerce')
    df = df[df[df.columns[1]].notnull()]
    df[df.columns[1]].plot(kind='hist', bins=10)

    # Add a title and label the axes
    plt.title('Histogram of ' + i[1])
    plt.xlabel('Value')
    plt.ylabel('Frequency')

    # Show the plot
    plt.show()
#     except:
#         pass
```
